# 模型对比
聚合不同模型与种子的 P/R/F1，生成表格与条形图。

```python
import json, os, glob
import pandas as pd
import matplotlib.pyplot as plt
root = os.path.abspath(os.path.join(os.getcwd(), '..'))
results_dir = os.path.join(root, 'results')
rows = []
for p in glob.glob(os.path.join(results_dir, 'results_seed_*_*.json')):
    with open(p, 'r', encoding='utf-8') as f:
        data = json.load(f)
    name = os.path.basename(p)
    m = data['test']
    rows.append({'file': name, **m})
df = pd.DataFrame(rows)
print(df)
df[['precision','recall','f1','accuracy']].plot(kind='bar'); plt.show()
```
