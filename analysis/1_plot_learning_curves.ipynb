# 学习曲线可视化
从 results/ 加载训练/验证损失与准确率，绘制随 epoch 的变化。

```python
import json, os, glob
import matplotlib.pyplot as plt
root = os.path.abspath(os.path.join(os.getcwd(), '..'))
results_dir = os.path.join(root, 'results')
paths = glob.glob(os.path.join(results_dir, 'results_seed_*_*.json'))
for p in paths:
    with open(p, 'r', encoding='utf-8') as f:
        data = json.load(f)
    train_losses = [e['train']['loss'] for e in data['epochs']]
    val_losses = [e['val']['loss'] for e in data['epochs']]
    plt.figure(); plt.plot(train_losses, label='train'); plt.plot(val_losses, label='val');
    plt.title(os.path.basename(p)); plt.xlabel('epoch'); plt.ylabel('loss'); plt.legend(); plt.show()
```
